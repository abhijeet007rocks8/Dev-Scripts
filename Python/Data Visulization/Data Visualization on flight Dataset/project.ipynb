# **Visualizing Flight Data Set**

Visualizing the flight data set of the different airlines on the bases of the Source and Destination of the flight, Airline company, Route, Total stops, etc. 

**Libraries are used :**
- Pandas ->  for creating data frames and reading data from the csv file.
- Numpy ->  for creating arrays of the Source of the airline and airline companies from the flight dataset.
- Matplotlib -> for creating the different types of the graph (like: pie charts, histogram, etc) for comparing the data
- Seaborn ->  for creating the different types of bar charts

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb 

%matplotlib inline
```

## **Creating the Data frames**

Read the data from the CSV file and creating and storing the "flights" data frame and displaying the retrieving data from the csv file.

```python
flights = pd.read_csv('flights_data.csv')
print(flights.shape)   # it tells no of rows and columns in the file
flights.head()
```

## **Bar Charts**

A Chart that represent the categorical data with rectangular bars with their height or length that proportional to their values that to represent. Here we are creating the different types of bar charts on different parameters like source and no. of flights, Airline and Output, etc using seaborn libraries. 

Represent the Bar charts :
- in Horizontal and Vertical format
- with different colors of bars, etc.

```python
# Bar charts

sb.countplot(data=flights,x='Source')
#plt.xtickets(rotation=30)
plt.ylabel('Number of Flights',fontsize=12)
plt.xlabel("Source",fontsize=12)
```

```python
base_color=sb.color_palette()[0]
sb.countplot(data=flights,x='Source',color=base_color)
plt.xticks(rotation=30)
```

```python
base_color=sb.color_palette()[1]
gen_order = flights['Source'].value_counts().index
sb.countplot(data=flights,x='Source',color=base_color,order=gen_order)
```

```python
base_color=sb.color_palette()[2]
sb.countplot(data=flights,x='Airline',color=base_color)
```

```python
base_color=sb.color_palette()[2]
sb.countplot(data=flights,x='Airline',color=base_color)
plt.xticks(rotation=90)
```

```python
base_color=sb.color_palette()[2]
sb.countplot(data=flights,y='Airline',color=base_color)
```

```python
# Count Missing Data 

flights.isna().sum()
```

```python
na_counts = flights.isna().sum()
base_color=sb.color_palette()[0]
sb.barplot(na_counts.index.values, na_counts, color=base_color)
plt.xticks(rotation=90)
plt.ylabel('Number of missing values', fontsize=12)
```

## **Pie Charts**

A pie chart that represents the data in circular graph and the slices of the graph shows the relative size of the data. Here we are going to create different type of Bar charts based on the different types of parameters .

```python
# Pie chart 

#code for the pie chart seen below
sorted_counts = flights['Destination'].value_counts()
plt.pie(sorted_counts, labels=sorted_counts.index, startangle=90, counterclock = False)
plt.axis('square')
plt.title('Flight Destination\'s')
```

```python
sorted_counts = flights['Destination'].value_counts()
plt.pie(sorted_counts, labels=sorted_counts.index, startangle=90, counterclock = False, wedgeprops={'width':0.4})
plt.axis('square')
plt.title('Flight Destination\'s')
```

## **Histograms**

Histogram represents a frequency distribution by means of rectangles whose widths represent class intervals and whose areas are proportional to the corresponding frequencies: the height of each is the average frequency density for the interval. Histogram groups the numbers into the ranges and it is also used to show frequency distribution. Histogram is only used to plot the frequency of score occurrences in a continuous data set that has been divided into classes.

So here we are visualizing Histogram based on :
- the duration of the Flight from source to destination.
- cost of the airbus travelling, etc.

Here we are going to use matplotlib.pyplot libraries for visualizing the data in Histograms.

```python
# Histograms

plt.hist(data=flights, x='Duration(minutes)')
```

```python
plt.hist(data=flights, x='Price', bins=20)
```

```python
plt.hist(data=flights,x='Price',bins=20)
```

```python
bins = np.arange(0, flights['Price'].max()+1, 1200)
plt.hist(data = flights, x='Price',bins=bins)
plt.show()
```

```python
sb.distplot(flights['Price'])
```

```python
sb.distplot(flights['Price'],kde=False)
```

```python
bin_edges=np.arange(0, flights['Price'].max()+1, 1200)
sb.distplot(flights['Price'],bins = bin_edges,kde = False, hist_kws = {'alpha':1})
```
